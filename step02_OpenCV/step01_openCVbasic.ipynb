## 3. matplotlib 로 이미지 그리기

```python
imgBGR = cv2.imread('imglist/cat.bmp')

# cv2.imread() 함수로 read 한 영상의 색상 정보는 RGB가 아닌 BGR 순서로 변경
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

# x, y 좌표값 표현 여부
plt.axis('off')
plt.imshow(imgRGB)
plt.show()
```

```python
# 그레이스케일 영상 출력
imgGray = cv2.imread('imglist/cat.bmp', cv2.IMREAD_GRAYSCALE)

# plt.axis('off')
plt.imshow(imgGray, cmap='gray')
plt.show()
```

```python
# 두 개의 영상을 함께 출력
plt.subplot(121), plt.axis('off'), plt.imshow(imgRGB)
plt.show()
```

```python
plt.subplot(122), plt.axis('off'), plt.imshow(imgGray, cmap='gray')
plt.show()
```
